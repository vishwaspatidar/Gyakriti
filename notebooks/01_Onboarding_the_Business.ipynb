```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
def load_dataset(file_name: str) -> pd.DataFrame:
    raw_data_dir = Path("/Users/vishwashpatidar/Project Gyakriti/Gyakriti/data/raw")
    return pd.read_csv(raw_data_dir / file_name)
```

```python

customers = load_dataset("olist_customers_dataset.csv")
orders = load_dataset("olist_orders_dataset.csv")
items = load_dataset( "olist_order_items_dataset.csv")
payments = load_dataset( "olist_order_payments_dataset.csv")
review = load_dataset( "olist_order_reviews_dataset.csv")
geolocation = load_dataset( "olist_geolocation_dataset.csv")
products = load_dataset( "olist_products_dataset.csv")
seller = load_dataset( "olist_sellers_dataset.csv")
prod_cat = load_dataset ("product_category_name_translation.csv")
```

```python
raw_tables = {
    "customers": customers,
    "orders": orders,
    "order_items": items,
    "payments": payments,
    "reviews": review,
    "products": products,
    "sellers": seller,
    "geolocation": geolocation,
    "prod_cat": prod_cat,
}
```
